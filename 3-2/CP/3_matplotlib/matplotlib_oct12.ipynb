# Matplotlib 2

```python
x = np.linspace(0, 20, 10)
y1 = x
y2 = 2*x
y3 = 3*x
y4 = 4*x
y5 = 5*x

plt.plot(x, y1, label='y=x')
plt.plot(x, y2, label='y=2x')
plt.plot(x, y3, label='y=3x')
plt.plot(x, y4, label='y=4x')
plt.plot(x, y5, label='y=5x')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.legend(loc=(0.05,0.6), frameon=False) # loc=lower left conner location
plt.show()
```

```python
plt.plot(x, y1, ls=':', lw=1, label='y=x')
plt.plot(x, y2, ls='-.', lw=1.5, label='y=2x')
plt.plot(x, y3, ls='--', lw=2, label='y=3x')
plt.plot(x, y4, ls='-', lw=2.5, label='y=4x')
plt.plot(x, y5, ls=(0, (1,1)), lw=3, label='y=5x')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.legend()
plt.show()
```

```python
plt.plot(x, y1, ls=':', lw=1, marker='o', ms=5, label='y=x')
plt.plot(x, y2, ls='-.', lw=1.5, marker='+', ms=10, label='y=2x')
plt.plot(x, y3, ls='--', lw=2, marker='x', ms=10, label='y=3x')
plt.plot(x, y4, ls='-', lw=2.5, marker='s', ms=8, mfc='None', mew=2, label='y=4x')
plt.plot(x, y5, ls=(0, (1,1)), lw=3, marker='D', ms=8, label='y=5x')
#mfc='None', mew = num
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.legend(frameon=False)
plt.show()
```

# Matplotlib 3

```python
plt.plot(x, y1, c='0.2', ls=':', lw=2, marker='o', ms=5, label='y=x')
plt.plot(x, y2, ls='-.', lw=1.5, marker='+', ms=10, label='y=2x')
plt.plot(x, y3, ls='--', lw=2, marker='x', ms=10, alpha=0.5, label='y=3x')
plt.plot(x, y4, ls='-', lw=2.5, marker='s', ms=8, mfc='0.8', mew=2, label='y=4x')
plt.plot(x, y5, lw=1.5, marker='D', c='r', ms=8, mfc='w', mec='C3', label='y=5x')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.legend(frameon=False)
plt.show()
```

```python
plt.plot(x, y1, 'ko:', lw=2, ms=5, label='y=x')
plt.plot(x, y2, '+-.', lw=1.5, ms=10, label='y=2x')
plt.plot(x, y3, 'x--', lw=2, ms=10, alpha=0.5, label='y=3x')
plt.plot(x, y4, 's-', lw=2.5, ms=8, mfc='0.8', mew=2, label='y=4x')
plt.plot(x, y5, 'rD-', lw=1.5, ms=8, mfc='w', mec='C3', label=r'$y=5\cdot x$')
# fmt replaces c, marker, ls. 
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.legend(frameon=False)
plt.show()
```

```python
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)
err = 0.1+0.2*x
plt.errorbar(x, y, yerr=err, fmt='-o')
plt.show()
```

```python
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)
lowerr = 0.4*err
uperr = err
asymm_err = [lowerr, uperr]
plt.errorbar(x, y, yerr=asymm_err, errorevery=2,
             ecolor='C3', elinewidth=1, capsize=4, capthick=1, fmt='o')
plt.show()
```

# Matplotlib 4

```python
x = np.arange(0.1, 4, 0.5)
y = np.exp(-x)
err = 0.1+0.2*x
plt.errorbar(x, y, yerr=err, 
             fmt='-o', elinewidth=1, capsize=4, capthick=1)
plt.xlim(0, 3.7)
plt.ylim(-0.9, 1.1)
plt.show()
```

```python
plt.plot(x, y1, 'ko:', lw=2, ms=5, label='y=x')
plt.plot(x, y2, '+-.', lw=1.5, ms=10, label='y=2x')
plt.plot(x, y3, 'x--', lw=2, ms=10, alpha=0.5, label='y=3x')
plt.plot(x, y4, 's-', lw=2.5, ms=8, mfc='0.8', mew=2, label='y=4x')
plt.plot(x, y5, 'rD-', lw=1.5, ms=8, mfc='w', mec='C3', label=r'$y=5\cdot x$')
plt.xlabel('x', fontsize=15)
plt.ylabel('y', fontsize=15)
plt.ylim(0, 100)
plt.xlim(0, 20)
#plt.yticks([])
plt.yticks(np.arange(0, 101, 50), [r'$\alpha$', r'$\beta$', r'$\gamma$'])
plt.xticks(np.arange(0, 21, 5))
plt.legend(frameon=False)
plt.show()
```

```python
x = 2**np.linspace(0, 11, 10000)
y = x**(-2)
plt.plot(x, y)
plt.xlim(1, 2**9)
plt.ylim(0.00001, 1)
plt.yscale('log')
plt.xscale('log', basex=2)
plt.grid()
plt.show()
```
