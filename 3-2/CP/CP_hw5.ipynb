```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as ticker
```

```python
mpl.rcParams.update({
    'font.family' : 'STIXGeneral',
    'mathtext.fontset' : 'stix',
    'font.size' : 14,
    'xtick.major.size' : 5 ,
    'xtick.minor.size' : 3 ,
    'xtick.major.width' : 1.2 ,
    'xtick.minor.width' : 1.2 ,
    'xtick.labelsize' : 13 ,
    'xtick.direction' : 'in' ,
    'xtick.top' : True ,
    'ytick.major.size' : 5 ,
    'ytick.minor.size' : 3 ,
    'ytick.major.width' : 1.2 ,
    'ytick.minor.width' : 1.2 ,
    'ytick.labelsize' : 13 ,
    'ytick.direction' : 'in' ,
    'xtick.major.pad' : 8 ,
    'xtick.minor.pad' : 8 ,
    'ytick.major.pad' : 5 ,
    'ytick.minor.pad' : 5 ,
    'ytick.right' : True ,
    'savefig.dpi' : 600 ,
    'axes.unicode_minus' : False ,
    'axes.linewidth' : 1.2 ,
    'lines.linewidth' : 1.5,
    'legend.frameon' : False,
    'legend.fontsize' : 15
})
```

```python
fig, ax = plt.subplots(2, 2, figsize=(7.0, 6.3))

x = np.linspace(0.01, 10)
ax[0,0].plot(x, 1-x/10, label=r'$y=1-\frac{1}{10}x$')
ax[0,0].plot(x, np.exp(-x), '--', label=r'$y=e^{-x}$')
ax[0,0].plot(x, x**(-2), ':', label=r'$y=x^{-2}$')
ax[0,0].set_ylim(0, 1)
ax[0,0].set_xlim(0, 10)
ax[0,0].legend(handlelength=0.8, labelspacing=0.1, handletextpad=0.5)
ax[0,0].set_ylabel('$y$')
ax[0,0].xaxis.set_major_locator(ticker.MultipleLocator(5))
ax[0,0].xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax[0,0].yaxis.set_major_locator(ticker.MultipleLocator(0.2))

ax[0,1].plot(x, 1-x/10, c='C0', label=r'$y=1-\frac{1}{10}x$')
ax[0,1].set_ylim(0, 1)
ax[0,1].set_xlim(0, 10)
ax[0,1].legend(handlelength=0.8, handletextpad=0.5)
ax[0,1].xaxis.set_major_locator(ticker.MultipleLocator(5))
ax[0,1].xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax[0,1].yaxis.set_major_locator(ticker.MultipleLocator(0.2))

ax[1,0].plot(x, np.exp(-x), '--', c='C1', label=r'$y=e^{-x}$')
ax[1,0].set_ylim(1e-4, 1)
ax[1,0].set_xlim(0, 10)
ax[1,0].set_yscale('log')
ax[1,0].legend(handlelength=0.8, handletextpad=0.5)
ax[1,0].set_xlabel('$x$')
ax[1,0].set_ylabel('$y$')
ax[1,0].xaxis.set_major_locator(ticker.MultipleLocator(5))
ax[1,0].xaxis.set_minor_locator(ticker.MultipleLocator(1))
ax[1,0].yaxis.set_major_locator(ticker.LogLocator(base=10))
ax[1,0].yaxis.set_minor_locator(ticker.NullLocator())

ax[1,1].loglog(2**x, (2**x)**(-2), ':', c='C2', label=r'$y=x^{-2}$')
ax[1,1].set_ylim(1e-6, 1)
ax[1,1].set_xlim(1, 2**10)
ax[1,1].legend(handlelength=0.8, handletextpad=0.5)
ax[1,1].set_xlabel('$x$')
ax[1,1].xaxis.set_major_locator(ticker.LogLocator(base=2, numticks=11))
ax[1,1].xaxis.set_major_formatter(ticker.LogFormatterMathtext(base=2))
ax[1,1].yaxis.set_major_locator(ticker.LogLocator(base=100))
ax[1,1].yaxis.set_minor_locator(ticker.LogLocator(base=100.0, subs=(0.1,)))
ax[1,1].yaxis.set_minor_formatter(ticker.NullFormatter())

plt.tight_layout(pad=0.5, w_pad=2, h_pad=1)
fig.text(0.01, 0.965, '(a)', fontsize=15)
fig.text(0.505, 0.965, '(b)', fontsize=15)
fig.text(0.01, 0.485, '(c)', fontsize=15)
fig.text(0.505, 0.485, '(d)', fontsize=15)
plt.savefig("test.pdf")
```
