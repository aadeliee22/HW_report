```python
import numpy as np
from time import time
import matplotlib.pyplot as plt
```

```python
# jacobi: update after every iteration. symmetry, less stable, slow
M = 100
V = 1
acc = 1e-4

phi = np.zeros((M+1, M+1))
phi[0, :] = V
phinew = np.zeros_like(phi)
d = 1
t = time()
while d > acc:
    for i in range (M+1):
        for j in range (M+1):
            if i==0 or i==M or j==0 or j==M:
                phinew[i,j] = phi[i,j] #exclude boudary when update
            else:
                phinew[i,j] = (phi[i+1,j]+phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4
    d = np.max(np.abs(phi-phinew))
    phi, phinew = phinew, phi #deep copy, swap
print("time:", time()-t)
```

```python
ax = plt.gca()
im = ax.imshow(phi)
cb = plt.colorbar(im)
```

```python
# gauss-seidel: update right away. no symmetry, stable, fast

phi = np.zeros((M+1, M+1))
phi[0, :] = V
d = 1
t = time()
while d > acc:
    d = 0
    for ii in range (M-1):
        for jj in range (M-1):
            i, j = ii+1, jj+1
            tmp = (phi[i+1,j]+phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4
            d = max(d, abs(phi[i,j]-tmp))
            phi[i,j] = tmp
print("time:", time()-t)
```

```python
ax = plt.gca()
im = ax.imshow(phi)
cb = plt.colorbar(im)
```

```python
# sor
phi = np.zeros((M+1, M+1))
phi[0, :] = V
acc = 1e-4
w = 0.9
a = 0.25*(1+w)
b = -w

d = 1
t = time()
while d > acc:
    d = 0
    for ii in range (M-1):
        for jj in range (M-1):
            i, j = ii+1, jj+1
            tmp = a*(phi[i+1,j]+phi[i-1,j] + phi[i,j+1] + phi[i,j-1]) + b*phi[i,j]
            d = max(d, abs(phi[i,j]-tmp))
            phi[i,j] = tmp
print("time:", time()-t)
ax = plt.gca()
im = ax.imshow(phi)
cb = plt.colorbar(im)
```

```python
# j + or
acc = 1e-3
w = 0.05

phi = np.zeros((M+1, M+1))
phi[0, :] = V
phinew = np.zeros_like(phi)

d = 1
t = time()
while d > acc:
    for i in range (M+1):
        for j in range (M+1):
            if i==0 or i==M or j==0 or j==M:
                phinew[i,j] = phi[i,j] 
            else:
                phinew[i,j] = (phi[i+1,j]+phi[i-1,j] + phi[i,j+1] + phi[i,j-1])/4
    phinew = (1+w)*phinew - w*phi
    d = np.max(np.abs(phi-phinew))
    print(d)
    phi, phinew = phinew, phi
print("time:", time()-t)
ax = plt.gca()
im = ax.imshow(phi)
cb = plt.colorbar(im)
```
