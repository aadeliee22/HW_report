## Nonlinear eq 2

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# newton
a = 1e-6
u = np.linspace(-0.99, 0.99, 200)
x = np.zeros(len(u))
for i in range (len(u)):
    x0 = 0
    for j in range (1000):
        d = (np.tanh(x0)-u[i])*np.cosh(x0)**2
        if np.abs(d)<a: break
        x0 = x0 - d
    x[i] = x0
plt.plot(u, x, 'b-')
```

## Minimization, local

```python
from scipy import optimize as opt
```

```python
def f(x):
    return x**4 + 3*(x-2)**3-15*x**2+1

x = np.linspace(-8, 5, 100)
plt.plot(x, f(x))

opt.minimize_scalar(f, method='Brent')
```

```python
opt.minimize_scalar(f, method='bounded', bounds=[0, 6])
```

```python
x = np.linspace(-5, 5, 100)
y = x
X, Y = np.meshgrid(x, y)
Z = opt.rosen(np.vstack([X.ravel(), Y.ravel()])).reshape((100, 100))

def reporter(p): #history
    global ps
    ps.append(p)
```

```python
x0 = np.array([4, -3])
ps = [x0]
opt.minimize(opt.rosen, x0, method='nelder-mead', callback=reporter)
#opt.minimize(opt.rosen, x0, method='Newton-CG', jac=opt.rosen_der, hess=opt.rosen_hess, callback=reporter)

ps = np.array(ps)
fig, ax = plt.subplots(1, 2, figsize=(12,4))
ax[0].contour(X, Y, Z, np.arange(10)**5)
ax[0].plot(ps[:,0], ps[:,1], 'r-.')
ax[1].semilogy(range(len(ps)), opt.rosen(ps.T))
```

```python
s = 1
a = 1e-6
phi = 2/(1+np.sqrt(5))

V = lambda r: (s/r)**6 - np.exp(-r/s)
x0 = 0.1*s
x3 = 5*s
x = np.array([x0, x3 + (x0 - x3) * phi, x0 - (x0 - x3) * phi, x3])
y = V(x)

while (x[3]-x[0])>a:
    if y[1] < y[2]:
        x[3], x[2] = x[2], x[1]
        x[1] = x[3] + (x[0] - x[3]) * phi
        y = V(x)
    else:
        x[0], x[1] = x[1], x[2]
        x[2] = x[0] - (x[0] - x[3]) * phi
        y = V(x)
print((x[0]+x[3])/2)
```

```python
x = np.linspace(-5, 5, 100)
y = 2.9*np.sin(1.5*x) + np.random.normal(size=100)
def func(x, a, b):
    return a*np.sin(b*x)
p, pcov = opt.curve_fit(func, x, y, p0=[2,2])
print(p)

plt.plot(x, y, 'b.')
plt.plot(x, func(x, *p), 'b-')
```
