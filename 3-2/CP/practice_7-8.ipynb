## Function

````python
a = 1
def vartest(a):
    a = a + 1
vartest(a)
print(a)
````

````python
a = 1
def vartest(a):
    a = a + 1
    return a
a = vartest(a)
print(a)
````

## Unpacking

````python
def change(a, b, c):
    return b, a, c
p = [5, 7, 9]
change(*p)
````

````python
def change(a, *args):
    return args[0], a, args[1]
change(*(1, 5, 9))
````

## Lambda
lambda arg1, arg2, $\cdots$ : return

````python
def add(a, b):
    return a+b
def mult(a, b):
    return a*b;
mylist=[add, mult]
print(mylist[0](3, 4), mylist[1](3, 4))
````

````python
mylist = [lambda a, b: a+b, lambda a, b: a*b] # better
print(mylist[0](3, 4), mylist[1](3, 4))
````

````python
class Calculator:
    def __init__(self): # constructor
        self.result = 0
    def adder(self, num):
        self.result += num
        return self.result
    
cal1 = Calculator()
cal2 = Calculator()

print(cal1.adder(3), cal2.adder(3))
print(cal1.adder(4), cal2.adder(7))
````
