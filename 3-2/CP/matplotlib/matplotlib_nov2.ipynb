```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
t = np.arange(0.01, 10.0, 0.01)
data1 = np.exp(t)
data2 = np.sin(2 * np.pi * t)
ax1 = plt.gca()
color = 'tab:red'
ax1.set_xlabel('time [s]')
ax1.set_ylabel('exp', c=color)
ax1.plot(t, data1, c=color)
ax1.tick_params(axis='y', labelcolor=color)


ax2 = ax1.twinx()
color = 'blue'
ax2.set_ylabel('sin', c=color)
ax2.plot(t, data2, c=color)
ax2.tick_params(axis='y', labelcolor=color)
```

```python
N = 100000
n_bin = 20
x = np.random.randn(N)
y = 0.4 * x + np.random.randn(N) + 5
fig, ax = plt.subplots(1, 2, figsize=(6, 4), sharey=True)
ax[0].hist(x, bins=n_bin, histtype='step', density=True, align='mid')
ax[1].hist(y, bins=n_bin, histtype='step', density=True, align='right')
ax[0].axvline(0)
ax[1].axvline(5)
```

```python
n = 1000000
s = np.random.pareto(2.0, n)+1
bins = np.logspace(1, 12, 12, base=2)
ax = plt.gca()
count, bins = np.histogram(s, bins=bins)
w = bins[1:-1]-bins[:-2]
x = bins[:10]
y = count[:10]/w/n
ax.loglog(x, y, 'k+', label='actual data')
ax.loglog(x, x**-3, 'k:', label=r'$x^{-3}$')
ax.set_xlim([1, 2048])
ax.set_xlabel('x')
ax.set_ylabel('P(x)')
ax.legend(frameon=False)
```

```python
ax = plt.gca()
x = np.random.randn(100000)
y = 0.4 * x + np.random.randn(100000) + 5
counts, xedges, yedges, im = ax.hist2d(x, y, bins=40, density=True)
cb = plt.colorbar(im, ax=ax)
cb.set_label("probability density")
```

```python
y, x = np.mgrid[-3.0:3.01:0.05,-6.0:6.01:0.15]
z = (1 - 0.5*x + x**5 + y**3) * np.exp(-x**2 - y**2)
z = z[:-1, :-1]
z_min, z_max = -np.abs(z).max(), np.abs(z).max()
ax = plt.gca()
c = ax.pcolor(x, y, z, cmap='Spectral', vmin=z_min, vmax=z_max)
#c = ax.pcolormesh(x, y, z, cmap='RdBu', vmin=z_min, vmax=z_max)
ax.axis([x.min(), x.max(), y.min(), y.max()])
cb = plt.colorbar(c, ax=ax)
```

```python
data = np.clip(np.random.randn(250, 250), -1, 1)
ax = plt.gca()
cax = ax.imshow(data, interpolation='nearest', cmap='coolwarm')
cbar = plt.colorbar(cax, ticks=[-1, -0.5, 0, 0.5, 1])
cbar.ax.set_yticklabels(['< -1', '-0.5', '0', '0.5', '> 1'])
```

```python
ax = plt.gca()
cax = ax.imshow(data, interpolation='nearest', cmap='afmhot')
ax.invert_yaxis()
cbar = plt.colorbar(cax, ticks=[-1, 0, 1], orientation='horizontal')
cbar.ax.set_xticklabels(['Low', 'Medium', 'High'])
```
