```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.integrate as sint
```

```python
# nonlinear pendulum: \ddot{\theta} = -g/l \sin\theta
g = 9.8
l = 0.2
def f(s, t):
    th, om = s
    fth = om
    fom = -g/l*np.sin(th)
    return np.array([fth, fom])
```

```python
def Euler(f, x, t, h):
    return x + h*f(x,t)
def RK2(f, x, t, h):
    k1 = h*f(x, t)
    k2 = h*f(x+0.5*k1, t+0.5*h)
    return x+k2
def RK4(f, x, t, h):
    k1 = h*f(x, t)
    k2 = h*f(x+0.5*k1, t+0.5*h)
    k3 = h*f(x+0.5*k2, t+0.5*h)
    k4 = h*f(x+k3, t+h)
    return x+ (k1+2*k2+2*k3+k4)/6.
```

```python
N = 10000
t = np.linspace(0, 20, N)
s = np.zeros((N,2))
s[0,0] = np.pi/200
h = t[1]-t[0] #need very very small h

for i in range (N-1):
    s[i+1] = Euler(f, s[i], t[i], h)
plt.plot(t, s[:,0])
print('doomed')
```

```python
def leapfrog(f, x, xm, t, h): #xm = x(t+0.5*h)
    x1 = x + h*f(xm, t+0.5*h)
    xm1 = xm + h*f(x1, t+h)
    return x1, xm1
```

```python
N = 1000
t = np.linspace(0, 10, N)
s = np.zeros((N, 2))
s[0,0] = np.pi/5
h = t[1]-t[0]
sm = np.zeros_like(s)
sm[0] = RK4(f, s[0], t[0], 0.5*h) #making midpoint
for i in range (N-1):
    s[i+1], sm[i+1] = leapfrog(f, s[i], sm[i], t[i], h)
E = 0.5*l**2*s[:,1]**2 + g*l*(np.cos(s[0,0])-np.cos(s[:,0]))
plt.plot(t, E)
```

```python
from math import sin,pi
from numpy import empty,array,arange
from pylab import plot,show

g = 9.8
l = 0.1
theta0 = 179*pi/180

a = 0.0
b = 10.0
N = 100          # Number of "big steps"
H = (b-a)/N      # Size of "big steps"
delta = 1e-8     # Required position accuracy per unit time

def f(r):
    theta = r[0]
    omega = r[1]
    ftheta = omega
    fomega = -(g/l)*sin(theta)
    return array([ftheta,fomega],float)

tpoints = arange(a,b,H)
thetapoints = []
r = array([theta0,0.0],float)

# Do the "big steps" of size H
for t in tpoints:

    thetapoints.append(r[0])

    # Do one modified midpoint step to get things started
    n = 1
    r1 = r + 0.5*H*f(r)
    r2 = r + H*f(r1)

    # The array R1 stores the first row of the
    # extrapolation table, which contains only the single
    # modified midpoint estimate of the solution at the
    # end of the interval
    R1 = empty([1,2],float)
    R1[0] = 0.5*(r1 + r2 + 0.5*H*f(r2))

    # Now increase n until the required accuracy is reached
    error = 2*H*delta
    while error>H*delta:

        n += 1
        h = H/n

        # Modified midpoint method
        r1 = r + 0.5*h*f(r)
        r2 = r + h*f(r1)
        for i in range(n-1):
            r1 += h*f(r2)
            r2 += h*f(r1)

        # Calculate extrapolation estimates.  Arrays R1 and R2
        # hold the two most recent lines of the table
        R2 = R1
        R1 = empty([n,2],float)
        R1[0] = 0.5*(r1 + r2 + 0.5*h*f(r2))
        for m in range(1,n):
            epsilon = (R1[m-1]-R2[m-1])/((n/(n-1))**(2*m)-1)
            R1[m] = R1[m-1] + epsilon
        error = abs(epsilon[0])

    # Set r equal to the most accurate estimate we have,
    # before moving on to the next big step
    r = R1[n-1]

# Plot the results
plot(tpoints,thetapoints)
plot(tpoints,thetapoints,"b.")
show()
```

```python
def f(r): 
    rinv = np.sum(r**2)**(-1.5)
    return -r*rinv # -\vec{r}/|r|^3
def verlet(f, x, p, h): #velocity verlet
    p += 0.5*h*f(x)
    x += h*p
    p += 0.5*h*f(x)
    return x, p

vi = 1.25
x0 = np.array([1, 0])
p0 = np.array([0, vi])

ecc = 1 - vi**2
T = 2*np.pi*(1+ecc)**(-1.5)
h = 0.001
t = np.arange(0, min(T, 100), h)

x = np.zeros((t.size, 2))
p = np.zeros_like(x)
x[0], p[0] = x0, p0

for i in range (t.size-1):
    x[i+1], p[i+1] = verlet(f, x[i], p[i], h)

plt.plot(x[:,0], x[:,1], lw=1)
plt.plot(0,0, 'ko')
plt.axis('equal')
```

```python
#energy
K = 0.5 * np.sum(p**2, axis=1)
V = -1/np.sqrt(np.sum(x**2, axis=1))
E0 = 0.5*vi**2 - 1
plt.plot(t, K+V)
plt.ylim(0.9*E0, 1.1*E0)
```

```python
#angular momentum
L0 = vi
L = x[:,0]*p[:,1] - x[:,1]*p[:,0]
plt.plot(t, L)
plt.ylim(0.9*L0, 1.1*L0)
```
