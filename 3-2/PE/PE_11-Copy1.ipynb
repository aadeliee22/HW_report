```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
import matplotlib.ticker as ticker
import matplotlib as mpl
mpl.rcParams.update({
    'font.family' : 'STIXGeneral',
    'mathtext.fontset' : 'stix',
    'xtick.direction' : 'in' ,
    'xtick.labelsize' : 13 ,
    'xtick.top' : True ,
    'ytick.direction' : 'in' ,
    'ytick.labelsize' : 13 ,
    'ytick.right' : True ,
    'axes.labelsize' : 16,
    'legend.frameon' : False,
    'legend.fontsize' : 13,
    'legend.handlelength' : 1.5,
    'savefig.dpi' : 600, 
    'savefig.bbox' : 'tight'
})
```

```python
B = np.array([417,382,341,308,281,233,185,615])*1e-3
r = np.array([[124.8,155.4,200,140.1],
              [126.9,154.7,200.5,141.5],
              [128,153.2,200.4,140.7],
              [130,152.2,200.4,141.1],
              [130.6,150.8,200.5,141.5],
              [132.6,149.5,199.9,141.6],
              [134.8,147.6,199,141.8],
              [120.4,159.7,199,141.6]]) #computer unit

B0 = np.array([615, 417, 382, 341, 308, 281, 254, 233, 216, 199, 185, 174, 162])*1e-3
distance = (np.arange(13)+40)*1e-3
```

```python
# err. r
print((143.9-137.2)/2, (202.4-196.6)/2)
```

```python
d = r[:,1]**2 - r[:,0]**2
D = r[:,2]**2 - r[:,3]**2
dbyD = d/D
err_rout = 2.9
err_rin = 3.35
err_D = 2/D * np.sqrt(err_rin**2*(r[:,1]**2+r[:,0]**2+r[:,3]**2*dbyD**2) + dbyD**2*r[:,2]**2*err_rout**2)

def func(x, a, b): return a*x + b
def Rsquare(list_e, list_t):
    return 1 - np.sum((list_e - list_t)**2) / np.sum((list_e - np.average(list_e))**2)
```

```python
p,pcov = curve_fit(func, B, dbyD)

plt.errorbar(B, dbyD, yerr = err_D, marker = 'o', ms=5, c = 'b', ecolor='k', capsize=3, ls="",label='experiment')
#plt.plot(6.84*1e-3,0,'bo')
x = np.linspace(0, np.max(B))
txt = r'$%.2f x + %.2f$' %(p[0],p[1])
plt.plot(x, func(x, *p), 'b--', label=txt)
plt.legend()
plt.xlabel('B [T]')
plt.ylabel(r'$\delta/\Delta$')

print(p, np.diag(pcov), Rsquare(dbyD, func(B, *p)))

plt.savefig('6_1.png')
```

```python
h = 6.6261*1e-34
c = 2.9979*1e8
mu = 1.456
t = 0.003
me = 9.10938*1e-31
e = 1.602177*1e-19
muB0 = e*h / (4*np.pi*me)
muB = p[0]*h*c/(4*mu*t)
muB_err = np.diag(pcov)[0]*h*c/(4*mu*t)
print(muB0, muB, muB_err)
print("%.8e +- %.8e"%(muB*4*np.pi/h, muB_err*4*np.pi/h))
print(muB/muB0*100, muB_err/muB0*100)
```

```python
plt.plot(distance, B0, 'go')
plt.ylabel('B [T]')
plt.xlabel('Magnet distance [m]')

#plt.savefig('6_0.png')
```
