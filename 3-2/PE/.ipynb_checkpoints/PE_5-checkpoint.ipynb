```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.optimize import curve_fit
import pandas as pd
from sklearn.kernel_ridge import KernelRidge
```

```python
def Rsquare(list_e, list_t):
    return 1 - np.sum((list_e - list_t)**2) / np.sum((list_e - np.average(list_e))**2)
def chisquare(list_e, list_t):
    return np.sum((list_e - list_t)**2/list_t)
def Gtest(list_e, list_t, cut):
    return 2*np.sum(list_t[list_t>cut]-list_e[list_t>cut]+list_e[list_t>cut]*np.log(list_e[list_t>cut]/list_t[list_t>cut]))
```

```python
import matplotlib as mpl
mpl.rcParams.update({
    'font.family' : 'STIXGeneral',
    'mathtext.fontset' : 'stix',
    'xtick.direction' : 'in' ,
    'xtick.labelsize' : 11.5 ,
    'ytick.direction' : 'in' ,
    'ytick.labelsize' : 11.5 ,
    'axes.labelsize' : 16,
    'legend.frameon' : True,
    'savefig.dpi' : 600 
})
```

```python
t = 0.001
l = 0.02
A = 10**(-5)
e = 1.60217662 * 10**(-19) #e, eV = @ C, J
k_B = 1.3806 * 10**(-23)
R_p0 = 35
R_n0 = 37.3

Ierr = 1 #mA
Verr = 0.05 #mV
Berr = 5 #mT
Terr = 1 #*C

lb1=['th', 'I', 'B', 'Temp', 'V_p', 'V_n']
mk = ['o', '^', 's', 'D', 'v']
clr = ['b', 'r', 'g']

p = pd.read_excel('PE5.xlsx', names=lb1)
```

```python
def func_1(x, a, b):
    return a*x+b
def func_2(x, a, c):
    return a*(x)**2 + c
def func_3(x, a, b, c, d):
    return a*np.abs(x-b)+c*(x-b)+d
```

```python
fig, ax = plt.subplots(2, 1, figsize=(6, 9))

p1 = p[p.th==1]
popt1, pcov1 = curve_fit(func_1, p1.I, p1.V_p)
popt2, pcov2 = curve_fit(func_1, p1.I, p1.V_n)
print(popt1, np.sqrt(np.diag(pcov1)), Rsquare(p1.V_p, func_1(p1.I, *popt1)))
print(popt2, np.sqrt(np.diag(pcov2)), Rsquare(p1.V_n, func_1(p1.I, *popt2)))

for i in range (2):
    ax[0].plot(p1.I, p1[lb1[4+i]], marker=mk[i], ms=8, mfc='w', c = clr[i], lw=0, label=lb1[4+i])
txtst1 = 'y=%.2f x %.2f' % (popt1[0], popt1[1])
txtst2 = 'y=%.2f x + %.2f' % (popt2[0], popt2[1])
ax[0].plot(p1.I, func_1(p1.I, *popt1), 'b-.', lw = 1, label=txtst1)
ax[0].plot(p1.I, func_1(p1.I, *popt2), 'r--', lw = 1, label=txtst2)
ax[0].set_xlabel('I [mA]')
ax[0].set_ylabel('$V_H$ [mV]')
ax[0].legend(loc='center right')


p4 = p[p.th==4]
popt1, pcov1 = curve_fit(func_1, p4.B, p4.V_p)
popt2, pcov2 = curve_fit(func_1, p4.B, p4.V_n)
print(popt1, np.sqrt(np.diag(pcov1)), Rsquare(p4.V_p, func_1(p4.B, *popt1)))
print(popt2, np.sqrt(np.diag(pcov2)), Rsquare(p4.V_n, func_1(p4.B, *popt2)))

for i in range (2):
    ax[1].plot(p4.B, p4[lb1[4+i]], marker=mk[i], ms=8, mfc='w', c = clr[i], lw=0, label=lb1[4+i])
txtst1 = 'y=%.2f x %.2f' % (popt1[0], popt1[1])
txtst2 = 'y=%.2f x + %.2f' % (popt2[0], popt2[1])
ax[1].plot(p4.B, func_1(p4.B, *popt1), 'b-.', lw = 1, label=txtst1)
ax[1].plot(p4.B, func_1(p4.B, *popt2), 'r--', lw = 1, label=txtst2)
ax[1].set_xlabel('B [mT]')
ax[1].set_ylabel('$V_H$ [mV]')
ax[1].legend(loc='center right')
plt.savefig('5_1.png', bbox_inches='tight')
plt.show()

R_H_p = popt1[0]*t/0.03
R_H_n = -popt2[0]*t/0.03
R_H_perr = np.sqrt((t*np.sqrt(np.diag(pcov1))[0]/0.03)**2 + (t*popt1[0]*Ierr/30)**2)
R_H_nerr = np.sqrt((t*np.sqrt(np.diag(pcov2))[0]/0.03)**2 + (t*popt2[0]*Ierr/30)**2)
print("R_Hp =", R_H_p, R_H_perr)
print("R_Hn =", R_H_n, R_H_nerr)
print("mu_p =", l/(R_p0*A)*R_H_p, l/(R_p0*A)*R_H_perr)
print("mu_n =", l/(R_n0*A)*R_H_n, l/(R_n0*A)*R_H_nerr)
print("p =", 1/(e*R_H_p), R_H_perr/(e*R_H_p**2))
print("n =", 1/(e*R_H_n), R_H_nerr/(e*R_H_n**2))
```

```python
p2 = p[p.th==2]
popt1, pcov1 = curve_fit(func_2, p2.B, p2.V_p/np.min(p2.V_p), p0=[0.0001, 1])
popt2, pcov2 = curve_fit(func_2, p2.B, p2.V_n/np.min(p2.V_n), p0=[0.0001, 1])
print(popt1, np.sqrt(np.diag(pcov1)), Rsquare(p2.V_p/np.min(p2.V_p), func_2(p2.B, *popt1)))
print(popt2, np.sqrt(np.diag(pcov2)), Rsquare(p2.V_n/np.min(p2.V_n), func_2(p2.B, *popt2)))
txtst1 = 'y=%.2e $x^2$ + %.2f' % (popt1[0]**2, popt1[1])
txtst2 = 'y=%.2e $x^2$ + %.2f' % (popt2[0]**2, popt2[1])
for i in range (2):
    plt.plot(p2.B, p2[lb1[4+i]]/np.min(p2[lb1[4+i]]), marker=mk[i], ms=8, mfc='w', c = clr[i], lw=0, label=lb1[4+i])
plt.plot(p2.B, func_2(p2.B, *popt1), 'b-.', lw = 1, label=txtst1)
plt.plot(p2.B, func_2(p2.B, *popt2), 'r--', lw = 1, label=txtst2)
plt.xlabel('B [mT]')
plt.ylabel('$V/V_0$')
plt.legend(loc='upper left')
plt.savefig('5_3.png', bbox_inches='tight')
plt.show()
print("mu_p =", 1000*np.sqrt(popt1[0]), np.sqrt(np.diag(pcov1))[0]/(2*np.sqrt(popt1[0])))
print("mu_n =", 1000*np.sqrt(popt2[0]), np.sqrt(np.diag(pcov2))[0]/(2*np.sqrt(popt2[0])))
```

```python
p3 = p[p.th==3]
p3.Temp = p3.Temp+273
popt1, pcov1 = curve_fit(func_3, 1/p3.Temp, np.log(l*30/(A*p3.V_p)), p0 = [1767, 0.0028, 1482, 3.5])
popt2, pcov2 = curve_fit(func_3, 1/p3.Temp, np.log(l*30/(A*p3.V_n)), p0 = [1166, 0.00275, 833, 4])
print(popt1, np.sqrt(np.diag(pcov1)), Rsquare(np.log(l*30/(A*p3.V_p)), func_3(1/p3.Temp, *popt1)))
print(popt2, np.sqrt(np.diag(pcov2)), Rsquare(np.log(l*30/(A*p3.V_n)), func_3(1/p3.Temp, *popt2)))

for i in range (2):
    err_T = Terr/p3.Temp**2
    err_lns = np.sqrt((Ierr/30)**2+(Verr/p3[lb1[4+i]])**2)
    plt.errorbar(1/p3.Temp, np.log(l*30/(A*p3[lb1[4+i]])), xerr = err_T, yerr = err_lns, 
                 marker=mk[i], ms=5, mfc='w', c = clr[i], capsize=3, ls="", label=lb1[4+i])
txtst1 = 'slope: -%.2f' % (popt1[0]-popt1[2])
txtst2 = 'slope: -%.2f' % (popt2[0]-popt2[2])
plt.plot(np.linspace(0.002421, popt1[1]), func_3(np.linspace(0.002421, popt1[1]), *popt1), lw = 1, c='b', label = txtst1)
plt.plot(np.linspace(0.002421, popt2[1]), func_3(np.linspace(0.002421, popt2[1]), *popt2), lw = 1, c='r', label = txtst2)
plt.plot(np.linspace(popt1[1], 0.003339), func_3(np.linspace(popt1[1], 0.003339), *popt1), "b--", lw = 1)
plt.plot(np.linspace(popt2[1], 0.003339), func_3(np.linspace(popt2[1], 0.003339), *popt2), "r--", lw = 1)
plt.xlabel('1/T [K$^{-1}$]')
plt.ylabel('$\ln\sigma$')
plt.legend(loc='upper right')
plt.savefig('5_4.png', bbox_inches='tight')
plt.show();

print(2*k_B*(popt1[0]-popt1[2])/e, 2*k_B*(np.sqrt(np.diag(pcov1))[0]+np.sqrt(np.diag(pcov1))[2])/e)
print(2*k_B*(popt2[0]-popt2[2])/e, 2*k_B*(np.sqrt(np.diag(pcov2))[0]+np.sqrt(np.diag(pcov2))[2])/e)
```

```python
p3 = p[p.th==3]
for i in range (2):
    err_T = Terr/p3.Temp**2
    err_lns = np.sqrt((Ierr/30)**2+(Verr/p3[lb1[4+i]])**2)
    plt.errorbar(1/p3.Temp, np.log(l*30/(A*p3[lb1[4+i]])), xerr = err_T, yerr = err_lns, 
                 marker=mk[i], ms=5, mfc='w', c = clr[i], capsize=3, ls="", label=lb1[4+i])
ypred = KernelRidge(kernel='rbf', alpha = 9.5e-12).fit(np.array(1/p3.Temp).reshape(-1,1), 
                         np.array(np.log(l*30/(A*p3.V_p))).reshape(-1,1)).predict(np.array(1/p3.Temp).reshape(-1,1))
plt.plot(1/p3.Temp, ypred, 'b', lw=1)
```
