```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
q_e = 1.60217662 * 10**(-19) # C
m_e = 9.10938356 * 10**(-31) # kg
h = 6.62607015 * 10**(-34) # J s

alpha2 = (3.5*10**(-3)/5)**2
R = 8 * (4*np.pi*10**(-7)) * 154 * 5 / (np.sqrt(125) * (3.5*10**(-3)))
L = 0.135 # m
d = np.array([2.13 * 10**(-10), 1.23 * 10**(-10)])

p1 = pd.read_excel('PE2_1_em.xlsx', names=['trial', 'r', 'V', 'I_in', 'I_out', 'I', 'V_err', 'I_err'])
p2 = pd.read_excel('PE2_2_ediffract.xlsx', names=['V', 'nth', 'D_in', 'D_out', 'D', 'D_err'])
print(R)
```

```python
print(q_e/m_e*10**(-11))
```

## 2.1. e/m

```python
len(p1)
```

```python
for j in range (2, 6):
    i = j * 0.01
    plt.errorbar(p1.V[p1.r==i], p1.I[p1.r==i], yerr=p1.I_err[p1.r==i], 
                 marker=".", linestyle="", capsize=3, label=j)
plt.xlabel('V [V]', fontsize=15)
plt.ylabel('I [A]', fontsize=15)
plt.legend()
plt.savefig('1-1.png', dpi=800)
plt.show()
```

```python
x = p1.r**2 * p1.I**2 * alpha2
poly0 = np.poly1d(np.polyfit(x, 2*p1.V, 1))
poly1 = np.poly1d(np.polyfit(x[x>0.8*10**(-9)], 2*p1.V[x>0.8*10**(-9)]/x[x>0.8*10**(-9)], 0))
chi0 = np.sum((2*p1.V-poly0(x))**2)
chi1 = np.sum((2*p1.V-poly1[0]*x)**2)
sx = np.var(x) * len(p1)**2
t = np.array([0, 1, 2, 3, 4])
print(poly0, format(2*np.sqrt(chi0/sx), ".4e"))
print(poly1, format(2*np.sqrt(chi1/sx), ".4e"))

for j in range (2, 6):
    i = j * 0.01
    plt.errorbar(p1.r[p1.r==i]**2 * p1.I[p1.r==i]**2 * alpha2 * 10**9, 
                 2*p1.V[p1.r==i], 
                 xerr=2*p1.I[p1.r==i]*p1.r[p1.r==i]**2 * p1.I_err[p1.r==i] * alpha2 * 10**9,
                 marker=".", linestyle="", capsize=3, label=i)
plt.plot(t, poly0(t*10**(-9)), 'k--')
plt.plot(t, poly1[0]*t*10**(-9), 'k-')
plt.plot(t, t*q_e/m_e*10**(-9), 'm-', label='e/m')
plt.xlabel('$r^2 B^2 \\times 10^{-9}$ [$m^2 T^2$]', fontsize=15)
plt.ylabel('$2V$ [V]', fontsize=15)
plt.ylim(0)
plt.xlim(0, 4)
plt.legend()
plt.savefig('1-2.png', dpi=800)
plt.show()

print(poly0/(q_e/m_e), format(2*np.sqrt(chi0/sx)/(q_e/m_e), ".4e"))
print(poly1/(q_e/m_e), format(2*np.sqrt(chi1/sx)/(q_e/m_e), ".4e"))
```

## 2.2. 

```python
plt.figure(figsize=(4.5, 4.5))
tt = [1.3 * 10**(-11), 2.6 * 10**(-11)]
for i in range (1,3):
    plt.errorbar(h/np.sqrt(2*m_e*q_e*p2.V[p2.nth==i]), p2.D[p2.nth==i]*d[i-1]/(2*L), yerr=p2.D_err[p2.nth==i]*d[i-1]/(2*L),
                 marker="o", linestyle="", capsize=3, label=i)
plt.plot(tt, tt, 'k-')
print(1-np.sum((h/np.sqrt(2*m_e*q_e*p2.V)-p2.D*d[p2.nth-1]/(2*L))**2)
      /np.sum((p2.D*d[p2.nth-1]/(2*L)-np.average(p2.D*d[p2.nth-1]/(2*L)))**2))
print(1-np.sum((h/np.sqrt(2*m_e*q_e*p2.V[p2.nth==1])-p2.D[p2.nth==1]*d[0]/(2*L))**2)
      /np.sum((p2.D[p2.nth==1]*d[0]/(2*L)-np.average(p2.D[p2.nth==1]*d[0]/(2*L)))**2))
print(1-np.sum((h/np.sqrt(2*m_e*q_e*p2.V[p2.nth==2])-p2.D[p2.nth==2]*d[1]/(2*L))**2)
      /np.sum((p2.D[p2.nth==2]*d[1]/(2*L)-np.average(p2.D[p2.nth==2]*d[1]/(2*L)))**2))
plt.legend()
plt.xlim(1.4*10**(-11), 2.5*10**(-11))
plt.ylim(1.4*10**(-11), 2.5*10**(-11))
plt.xlabel('$\lambda_{theory} [m^{-1}]$', fontsize=15)
plt.ylabel('$\lambda_{exp} [m^{-1}]$', fontsize=15)
plt.savefig('2-1.png', dpi=800)
plt.show()
```

```python
poly2 = np.poly1d(np.polyfit(1/np.sqrt(p2.V[p2.nth==1]), p2.D[p2.nth==1]*np.sqrt(p2.V[p2.nth==1]), 0))
poly3 = np.poly1d(np.polyfit(1/np.sqrt(p2.V[p2.nth==2]), p2.D[p2.nth==2]*np.sqrt(p2.V[p2.nth==2]), 0))
chi2 = np.sum((p2.D[p2.nth==1]-poly2[0]/np.sqrt(p2.V[p2.nth==1]))**2)
chi3 = np.sum((p2.D[p2.nth==2]-poly3[0]/np.sqrt(p2.V[p2.nth==2]))**2)
sx2 = np.var(1/np.sqrt(p2.V[p2.nth==1])) * len(p2)**2
sx3 = np.var(1/np.sqrt(p2.V[p2.nth==2])) * len(p2)**2
t = np.array([0, 0.014, 0.016, 0.018, 0.02])
print(poly2, np.sqrt(chi2/sx2))
print(poly3, np.sqrt(chi3/sx3))

for i in range (1,3):
    plt.errorbar(1/np.sqrt(p2.V[p2.nth==i])*100, p2.D[p2.nth==i], yerr=p2.D_err[p2.nth==i],
                marker="o", linestyle="", capsize=3, label=i)
plt.plot(t*100, poly2[0]*t, 'b-', label="k1")
plt.plot(t*100, poly3[0]*t, 'r-', label="k2")
plt.ylim(0)
plt.xlim(0,2)
plt.legend()
plt.xlabel('$1/\sqrt{V}\\times10^{-2}\; [V^{-1/2}]$', fontsize=15)
plt.ylabel('D [m]', fontsize=15)
plt.savefig('2-2.png', dpi=1200)
plt.show()
d1_exp = 2*L*h/(np.sqrt(2*m_e*q_e)*poly2[0])
d2_exp = 2*L*h/(np.sqrt(2*m_e*q_e)*poly3[0])
d1_err = 2*L*h*np.sqrt(chi2/sx2)/(np.sqrt(2*m_e*q_e)*poly2[0]**2)
d2_err = 2*L*h*np.sqrt(chi3/sx3)/(np.sqrt(2*m_e*q_e)*poly3[0]**2)
print("d1 =", d1_exp, "+/-", d1_err)
print("d2 =", d2_exp, "+/-", d2_err)
```

```python
h_1 = np.sqrt(2*m_e*q_e)/(2*L) * d1_exp * poly2[0]
h_2 = np.sqrt(2*m_e*q_e)/(2*L) * d2_exp * poly3[0]
h_1err = np.sqrt(2*m_e*q_e)/(2*L) * np.sqrt((poly2[0]*d1_err)**2+(np.sqrt(chi2/sx2)*d1_exp)**2)
h_2err = np.sqrt(2*m_e*q_e)/(2*L) * np.sqrt((poly3[0]*d2_err)**2+(np.sqrt(chi3/sx3)*d2_exp)**2)
print(h_1, h_1err)
print(h_2, h_2err)
print(np.average(h_1, h_2), np.average())
```
