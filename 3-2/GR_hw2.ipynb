```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib as mpl
import matplotlib.ticker as ticker
mpl.rcParams.update({
    'font.family' : 'STIXGeneral',
    'mathtext.fontset' : 'stix',
    'font.size' : 17,
    'xtick.major.size' : 5.5 ,
    'xtick.minor.size' : 3 ,
    'xtick.major.width' : 1.2 ,
    'xtick.minor.width' : 1.2 ,
    'xtick.labelsize' : 15 ,
    'xtick.direction' : 'in' ,
    'xtick.top' : True ,
    'ytick.major.size' : 5.5 ,
    'ytick.minor.size' : 3 ,
    'ytick.major.width' : 1.2 ,
    'ytick.minor.width' : 1.2 ,
    'ytick.labelsize' : 15 ,
    'ytick.direction' : 'in' ,
    'xtick.major.pad' : 8 ,
    'xtick.minor.pad' : 8 ,
    'ytick.major.pad' : 5 ,
    'ytick.minor.pad' : 5 ,
    'ytick.right' : True ,
    'savefig.dpi' : 600 ,
    'axes.unicode_minus' : False ,
    'axes.linewidth' : 1.2 ,
    'lines.linewidth' : 1.5,
    'legend.frameon' : False,
    'legend.fontsize' : 13,
    'legend.handlelength' : 1.5
})
```

```python
fig, ax = plt.subplots(1, 1, figsize=(6, 5))
x = np.linspace(0.001, 2.5)
x = np.float64(x)
for i in range (1, 10):
    a = -np.log(0.5*i)
    ax.plot(x, np.log(x/(0.5*i)), 'k--')
    ax.plot(x, -np.log(x/(0.5*i)), 'k--')
    y = np.maximum(np.log(x/(0.5*i)), -np.log(x/(0.5*i)))
    ax.fill_between(x, y, np.maximum(y, np.ones(len(x), dtype=np.float64)*0.1), color='#edde05', lw=0)

ax.plot(x, -np.log(x/(0.5*i)), 'k--', label='light rays')
t = np.linspace(-2, 2)
ax.plot(1.5*np.cosh(t), t, 'k-', lw=2, label='World line')
y = np.maximum(np.log(x), -np.log(x/3))
ax.fill_between(x, y, np.maximum(y, np.ones(len(x), dtype=np.float64)*0.65), color='#edde05', lw=0)

ax.axhline(0, c='k', lw=1.2)
ax.axis('equal', adjustable='datalim')
ax.set_ylim(-0.9, 0.9)
ax.set_xlim(0, 2.5)
ax.set(xlabel='X', ylabel='T')
ax.legend(bbox_to_anchor=(1, 1))
plt.savefig('metric.png', bbox_inches='tight')
```
